```python
import GEOparse
import pandas as pd
```

```python
# GSE184269
```

```python
def extract_name(record):
    record = record.__dict__["metadata"]["supplementary_file"][0].split("/")[-1].replace("_Grn.idat.gz", "")
    record = "_".join(record.split("_")[1:])
    return record
```

```python
geo = GEOparse.get_GEO(geo="GSE184269", destdir="./")
```

```python
mapper = {"naăżve_b": "naive_B",
         "naăżve_cd4": "naive_CD4",
         "naăżve_cd8": "naive_CD8",
         "granulocytes": "granulocytes",
         "monocytes": "monocytes", 
         "natural_killer_cells": "NK",
         "peripheral_blood": "peripheral_blood",
         "whole_blood": "whole_blood"}
```

```python
pheno = geo.phenotype_data
id_mapper = {name: extract_name(record) for name, record in geo.gsms.items()}

pheno = pheno[["source_name_ch1", "characteristics_ch1.0.donor id"]]
pheno.columns = ["POI", "donor"]
pheno["key"] = pheno.index.map(id_mapper)

pheno["POI"] = pheno["POI"].str.lower()
pheno["POI"] = pheno["POI"].apply(lambda x: x.replace(" ", "_"))
pheno["POI"] = pheno["POI"].map(lambda x: mapper.get(x))

pheno["donor"] = pheno["donor"].str.upper()

pheno["sample"] = pheno["POI"] + "_" + pheno["donor"]
pheno.to_csv("../data/methylation_sample_sheet.csv")
pheno
```
