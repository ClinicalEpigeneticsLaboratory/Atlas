```python
import pandas as pd
```

```python
epic = pd.read_csv("../data/external/epic.csv", index_col=0, skiprows=7)
epic 
```

```python
epic = epic[["Relation_to_UCSC_CpG_Island", "UCSC_CpG_Islands_Name", "UCSC_RefGene_Group", "CHR_hg38", "Start_hg38", "End_hg38", "Strand_hg38"]]
```

```python
epic = epic.loc[[name for name in epic.index if name.startswith("cg")]]
```

```python
epic[["Relation_to_UCSC_CpG_Island", "UCSC_CpG_Islands_Name"]] = epic[["Relation_to_UCSC_CpG_Island", "UCSC_CpG_Islands_Name"]].fillna("Opensea")
epic["UCSC_RefGene_Group"] = epic["UCSC_RefGene_Group"].fillna("-")
epic = epic.dropna()
epic
```

```python
epic[["Start_hg38", "End_hg38"]] = epic[["Start_hg38", "End_hg38"]].astype(int)
epic
```

```python
epic = epic[~epic.CHR_hg38.isin(["chrX", "chrY", "chrM"])]
epic 
```
